4. Filter DataFrame with Selected Columns

```python
df = df[['continent', 'location', 'date', 'total_cases', 'total_deaths',
         'gdp_per_capita', 'human_development_index']]
df
```

```python
df
```

5. Data Cleaning

```python
df.drop_duplicates()
```

b) Find missing values in all columns

```python
df.isnull().sum()
```

c) Remove all observation where continent column value is missing

```python
df.dropna(subset=["continent"])
```

d) Fill all missing values with 0

```python
df=df.fillna(0)
```

```python
df.isnull().sum()
```

6. Data Time format

a) Convert 'date' column to datetime forma

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.dtypes
```

b) Create a new column 'month' from 'date'

```python
df['month'] = df['date'].dt.month
```

```python
df
```

7. Data Aggregation

a) Find max value in all columns using groupby on 'continent'

```python
df_groupby= df.groupby('continent').max().reset_index()
df_groupby
```

8. Feature Enginneering

a) Create 'total_deaths_to_total_cases' feature

```python
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
```

```python
df_groupby['total_deaths_to_total_cases']*100
```

```python
df_groupby
```

9. Data Visualization

a) Univariate Analysis

```python
sns.set(style='whitegrid')
```

```python
sns.distplot(df["gdp_per_capita"])
```

b) Plot a Scatter Plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_groupby, x='gdp_per_capita', y='total_cases', hue='continent', palette='Set2', s=100)
plt.title('Total Cases vs GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Total Cases')
plt.show()
```

c) Plot pair plot on df_groupby dataset

```python
sns.pairplot(data=df_groupby)
```

d) plot a bar plot

```python
sns.catplot(data=df_groupby, x='continent', y='total_cases', kind='bar', height=6, aspect=1.5, palette='pastel')
```
